```python
# Step 1: Replace -1 with NaN to identify missing values
df = df.replace(-1, np.nan)

# Step 2: Create a mask for original missing values (-1)
missing_mask = df.isnull()

# Step 3: Define relationships for imputation
impute_relations = {
    'LotFrontage': ['Neighborhood'],
    'GarageYrBlt': ['YearBuilt'],
    'MasVnrArea': ['MasVnrType'],
    'BsmtQual': ['Foundation'],
    'FireplaceQu': ['Fireplaces'],
    'GarageArea': ['GarageType', 'GarageCars']
}

# Step 4: Impute using relationships
for col, related_cols in impute_relations.items():
    if col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            df[col] = df.groupby(related_cols)[col].transform(lambda x: x.fillna(x.median()))
        else:
            df[col] = df.groupby(related_cols)[col].transform(lambda x: x.fillna(x.mode()[0]))

# Step 5: Fallback imputation for remaining missing values
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype in ['int64', 'float64']:
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0])

print("Remaining missing values:", df.isnull().sum().sum())
plt.figure(figsize=(12, 8))
missing_counts = missing_mask.sum().sort_values(ascending=False)
sns.barplot(x=missing_counts.values, y=missing_counts.index, palette='viridis')
plt.title('Original Count of -1 Values by Column')
plt.xlabel('Number of -1 Values')
plt.ylabel('Columns')
plt.grid(axis='x', linestyle='--')
plt.show()

plt.figure(figsize=(14, 10))
sns.heatmap(missing_mask, cmap='viridis', cbar=False, yticklabels=False)
plt.title('Original Heatmap of -1 Values (Missing Data)')
plt.show()
```
